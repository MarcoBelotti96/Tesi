```python
def inquinante_year(data_filename):
    #Same function as the one used in the AirNormalizer module
    def normalize_timeseries(model, data_):
        n_preds = 300
        verbose = False
        dfs = []
        for i in range(n_preds):
            if(verbose and i % 10 == 0):
                print(f"Making prediction {i}")

            scrambled_rows = data_.sample(frac = 1).reset_index(drop=True)
            scrambled_rows.index = data_.index
            scrambled_rows.epoch = data_.epoch
            preds = pd.DataFrame(model.predict(scrambled_rows.drop("inquinante", 1)), columns=["inquinante"])
            preds.index = data_.index
            dfs.append(preds)
        predictions = pd.concat(dfs)
        return predictions.groupby(predictions.index).mean()
    
    def create_model_until_year(year):
        threshold_date = dt.datetime(year, 12, 31)
        data_year = data[data.index < threshold_date]
        x_train = data_year.drop("inquinante", 1)
        y_train = data_year.inquinante
        model = RandomForestRegressor(n_estimators = 300, max_depth = 20, max_features = 5, oob_score = True, random_state = 123)
        model.fit(x_train, y_train)
        y_pred_train = model.predict(x_train)
        mse = metrics.mean_squared_error(y_train, y_pred_train)
        rmse = np.sqrt(mse)
        print(f"Training MSE: {mse}")
        print(f"Training RMSE: {rmse}")
        print(f"Training OOB score: {model.oob_score_}")
        y_pred_test = model.predict(data[data.index > threshold_date].drop("inquinante", 1))
        mse_test = metrics.mean_squared_error(data[data.index > threshold_date].inquinante, y_pred_test)
        rmse_test = np.sqrt(mse_test)
        r2 = model.score(data[data.index > threshold_date].drop("inquinante", 1), data[data.index > threshold_date].inquinante)
        print(f"Testing MSE: {mse_test}")
        print(f"Testing RMSE: {rmse_test}")
        print(f"Testing R2: {r2}")
        normalized_timeseries = normalize_timeseries(model, data)

        return (model, normalized_timeseries)
    
    data = pd.read_csv(data_filename, parse_dates=["Data"], index_col="Data")
    
    inquinante = data_filename.split("_")[2].split(".")[0]
    print(f"*** Working with {inquinante} ***")
    
    std_model = AirNormalizer(data)
    print("Standard model performance:")
    print("Training:")
    print(std_model.evaluate_training())
    print("Testing:")
    print(std_model.evaluate_testing())
    
    normalized_datafilename = data_filename[:-4] + "_normalizzato.csv"
    print(f"Norm filename: {normalized_datafilename}")
    std_normalized_timeseries = pd.read_csv(normalized_datafilename, parse_dates=["Data"], index_col="Data")
    
    base_year = 2010 if inquinante != "O3" else 2014
    
    for i in range(base_year, 2020):
        print(f"Model created with data until year {i}")
        plt.clf() #clear the plot
        model, normalized_timeseries = create_model_until_year(i)
        y_max = max([max(normalized_timeseries.inquinante), max(std_normalized_timeseries.inquinante)]) * 1.1
        plt.plot(normalized_timeseries.index, normalized_timeseries.inquinante, color="green", label=f"Normalized with data until {i}")
        plt.plot(std_normalized_timeseries.index, std_normalized_timeseries.inquinante, color="red", label="Normalized with standard model")
        plt.ylim(0, y_max)
        plt.legend()
        plt.grid()
        plt.title(f"{inquinante} - Year {i}")
        plt.show()
        
for f in list(filter(lambda x: "normalizzato" not in x, glob.glob(path + "Mantova*.csv"))):
    #print(f"*** Working file: {f} ***")
    inquinante_year(f)
```
